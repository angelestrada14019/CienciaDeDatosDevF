##  weather

```python
weather=pd.read_csv('pydatadc_2018-tidy/data/weather.csv')
weather.head()
```

```python
weather_melt=pd.melt(
    weather,id_vars=['id','year','month','element'],
    var_name='day',
    value_name='temp'  
    )

weather_melt.head()
```

```python
weather_tidy=weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp',
    aggfunc=np.mean #si hay valor duplicado saca el promedio
)
weather_tidy.head()
```

```python
weather_tidy.reset_index()
```
